# Data preparation

```python
# setting the df to a subset of the df_original 

df = df_original [['coaster_name',
        #'Length', 'Speed', 
              'Location', 'Status', 
        #'Opening date',
       #'Type', 
        'Manufacturer',
        #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
        #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
        #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
df
```

```python
#Showing the data types of the data set

df.dtypes
```

```python
# Setting the data type of the column 'opening_date_clean' to to_datetime instead of the object type

df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
# exploring the data of the column 'opening_date_clean'

df['opening_date_clean']
```

```python
df.columns
```

```python
# Renaming some of the columns of the new data set

df = df.rename(columns= {
    'coaster_name': 'Coaster_Name',
    'year_introduced':'Year_Introduced',
    'opening_date_clean': 'Opening_Date',
    'speed_mph':'Speed_mph',
    'height_ft': 'Height_ft',
    'Inversions_clean': 'Inversions',
    'Gforce_clean': 'Gforce'
})

df
```

# Feature understanding

```python
# Count the number of roller coaster introduced per year

count_year = df['Year_Introduced'].value_counts().head(10)
count_year
```

```python
# Ploting the number of roller coaster introduced per year

ax =count_year.plot(kind = 'bar', title = 'top 10 years for roller coaster producing')
ax.set_xlabel('year introduced')
ax.set_ylabel('Count')
```

```python
# Showing the distribution of the speed using a histogram 

ax = df['Speed_mph'].plot(kind= 'hist', bins= 20, title = 'speed distribution', xlabel = 'Range', ylabel ='Speed')
```

```python
# Showing the distribution of the speed using a density distribution 

ax = df['Speed_mph'].plot(kind= 'kde',title = 'speed density distribution', xlabel = 'Range', ylabel ='Speed')
```

# Feature relationships

```python
df.head()
```

```python
df.plot(kind = 'scatter', x = 'Speed_mph', y = 'Height_ft')
plt.show()
```

```python
# Creating scatter plot between speed and height and setting the 'Year_Introduced' to be the hue
# Hue break the color by chosen criteria 

ax = sns.scatterplot(x = 'Speed_mph', y = 'Height_ft', data =df, hue = 'Year_Introduced' )
ax.set_title('Speed vs height')
plt.show()
```

```python
# Creating scatter plot between 'Year_Introduced','Speed_mph' in the X axis and 'Inversions','Gforce' in the y axis

ax = sns.pairplot(df, x_vars=['Year_Introduced','Speed_mph'], y_vars = ['Inversions','Gforce' ] , hue = 'Type_Main')
plt.show()
```

```python
# Creating 5*5 matrix of relation visualization between every numeric column and itself and other columns

ax = sns.pairplot(df, vars = ['Year_Introduced', 'Speed_mph', 'Height_ft','Inversions', 'Gforce'], hue = 'Type_Main')
plt.show()
```

```python
# calculating the correlation between every numeric column and itself and other columns

df_corr = df[['Year_Introduced', 'Speed_mph', 'Height_ft',
       'Inversions', 'Gforce']].dropna().corr()
df_corr
```

```python
# Visualizing the correlation table in heat map

sns.heatmap(df_corr, annot = True)
plt.show()
```

# Asking questions about data

```python
# Calculating the average and the mean for roller coaster speed for each Location

df.query("Location != 'Other'").groupby('Location')['Speed_mph'].agg(['mean','count'])
```

```python
# Ploting the top 10 locations with highest speed average

df.query("Location != 'Other'")\
.groupby('Location')['Speed_mph']\
.agg(['mean','count']).query('count >= 10')\
.sort_values('mean')['mean']\
.plot(kind = 'barh', xlabel ='Average speed', title = 'Average speed per location', color ='red')
plt.show()
```
